### Clase the variable continua uniforme greedy search - 05.09.23


```python
import random

U=4
L=0

inicio = 3
iteraciones = 500
x = inicio

i=0


def funcion(n):
  #Minimizar
  y = (n**5) - 10*(n**4) + 35*(n**3) - 50*(n**2) + 24*n
  return y

f = funcion(x)

#print(f)




while i < iteraciones:


    # #distribución uniforme
    # #--------------------------------------
    # des_x = random.random()
    # x1= x + (des_x - 0.5) * (U - L)

    #distribución normal
    #--------------------------------------
    des_x = random.normalvariate(0,1)
    x1= x + ((U - L)/6) *  des_x




    if(x1 >= L and x1 <= U):
        f1 = funcion(x1)
        if f1 < f:
            x = x1
            f = f1

    i = i+1

print("----------------")
print("Iteración:", i)
print("x*=", x)
print("f*=", f)
```

### Algoritmo de la mochila


```python
import random
import numpy as np


#Configuración inicial
#-----------------------------------
n=5     #No. de objetos (espacios) de la mochila
#Los siguientes arreglos se deben de adaptar en relación al valor de n
#sol[], NewSol[], W[], C[]

e=6     #Cantidad de elementos que puede haber en cada espacio (q)
nt=20  #No. de iteraciones
Ne_vecindad = 4 #No. elementos en la vecindad

sol=[0,0,0,0,0]     #solucion actual
NewSol = [0,0,0,0,0]  #nueva solución



#Datos del problema
#-----------------------------------
W = [1,2,5,7,7.5]
C = [1,2,3,6.5,7]

restriccion_pesos = 44
restriccion_cuota = 20


#Creación de otras variables
#-----------------------------------
#creamos vector de vecindad
vecindad = np.zeros((Ne_vecindad, n))

#creamos el vector que indica la calidad de las soluciones de la vecindad
#El primer valor nos indica si la solución es factible {0,1}
#Los otros valores son de F(x1) (peso) y F(x2) (cuota)
sol_evaluadas = np.zeros((Ne_vecindad, 3))


#Creación definiciones (funciones) para partir el problema
#-----------------------------------

def solucion_inicial():
    f=0
    r=0

    for k in range (100):


        for i in range(n):
            val = random.randint(0, e)
            #print(val)
            sol[i] = val

        f = validar_solucion(sol)

        if f==1:
            r=1
            break

    return r



def validar_solucion(nueva_solucion):

    suma_pesos = 0
    suma_cuota = 0

    for i in range(n):
        suma_pesos = suma_pesos + nueva_solucion[i] * W[i]
        suma_cuota = suma_cuota + nueva_solucion[i] * C[i]

    #print(suma_pesos)

    if suma_pesos <= restriccion_pesos and suma_cuota >= restriccion_cuota :
        return 1
    else:
        return 0

def evalua_vecindad():

    for k in range (Ne_vecindad):

        suma_pesos = 0
        suma_cuota = 0

        for i in range(n):
            suma_pesos = suma_pesos + vecindad[k][i] * W[i]
            suma_cuota = suma_cuota + vecindad[k][i] * C[i]

        #print(suma_pesos)

        sol_evaluadas[k][1] = suma_pesos
        sol_evaluadas[k][2] = suma_cuota

        if suma_pesos <= restriccion_pesos and suma_cuota >= restriccion_cuota :
            sol_evaluadas[k][0] = 1
        else:
            sol_evaluadas[k][0] = 0

    return 1


def genera_vecindad():


    for i in range (Ne_vecindad):
        NewSol = sol.copy()

        #print("\n", NewSol)

        modif = random.randint(0,n-1)
        agregar = random.randint(0,1)

        if agregar == 1:
            if NewSol[modif] < e:
                NewSol[modif] = NewSol[modif] +1
        else:
            if NewSol[modif] > 0:
                NewSol[modif] = NewSol[modif] -1
            #else:
             #   NewSol[modif] = NewSol[modif] +1

        #print(NewSol)

        vecindad [i] = NewSol.copy()



def seleccion_mejor():

    cm=0
    seleccionada = -1

    for i in range(Ne_vecindad):
        if sol_evaluadas[i][0] == 1:
            if sol_evaluadas[i][2] > cm:
                cm = sol_evaluadas[i][2]
                seleccionada = i

    return seleccionada



#Solución inicial y primera vecindad, iteración 1
#---------------------------------------------
print ("Vecindad vacia")
print (vecindad)

if( solucion_inicial() == 0):
    print("La solución inicial no es factible")

print("\nSolución inicial ->", sol)

print("\nGenera primera vecindad inmediata")
genera_vecindad()
print(vecindad)

evalua_vecindad()
print("\nEvalua vecindad")
print("[factibilidad, peso, cuota]")
print(sol_evaluadas)

ultima_mejor = seleccion_mejor()
if(ultima_mejor > -1):
    sol = vecindad[  ultima_mejor ].copy()



#Proceso iterativo
#---------------------------------------------

for t in range (nt-1):

    genera_vecindad()

    evalua_vecindad()

    ultima_mejor = seleccion_mejor()

    if(ultima_mejor > -1):
        sol = vecindad[  ultima_mejor ].copy()



#Imprimir resultados
#-----------------------------------
print("\nÚltima Vecindad inmediata")
print(vecindad)

print("\nEvalua vecindad")
print("[factibilidad, peso, cuota]")
print(sol_evaluadas)

print("\nSelecciona la mejor")
print (sol)

print("\nEvaluación de la mejor")
print (sol_evaluadas[ultima_mejor])

if(sol_evaluadas[ultima_mejor,0]==0):
    print ("No es factible!!!!!!!")
else:
    print("Factible!!!!!!!")



print("\nDatos del problema")
print("-----------------------------------------")
print("Vector pesos: ", W)
print("Vector de ganancia:", C)
print("Límite de pesos: ", restriccion_pesos)
print("Cuota mínima:", restriccion_cuota)
```

```python
import random

#La función tabu_searchtoma como entrada objetos
#(una lista de tuplas que representan los objetos con su valor y peso),
#capacidad_maxima(la capacidad máxima de la mochila),
#tamano_tabu(el tamaño de la lista tabú) y
#num_iteraciones(el número máximo de iteraciones).



def tabu_search(objetos, capacidad_maxima,  num_iteraciones, tiempo_listaT):
    # La función evaluar_solucion calcula el valor total y el peso total de una solución dada.
    #Si el peso total excede la capacidad máxima, devuelve 0; de lo contrario, devuelve el valor total.
    def evaluar_solucion(solucion):
        valor_total = 0
        peso_total = 0
        for i in range(len(solucion)):
            if solucion[i]:
                valor_total += objetos[i][0]
                peso_total += objetos[i][1]

        #print(solucion)
        #print("Valor tot= ", valor_total)
        #print("Peso tot= ", peso_total)

        if peso_total > capacidad_maxima:
            return 0
        return valor_total

    # La función generar_vecindarios genera todos los vecindarios posibles de una solución dada intercambiando un objeto seleccionado por uno no seleccionado.
    # Devuelve una lista de vecindarios.
    def generar_vecindarios(solucion):
        vecindarios = []
        no_objetos_mochila = len(solucion)

        for i in range(  no_objetos_mochila  ):


            vecindario = solucion.copy()
            if(  vecindario[i] == 0 ):
                vecindario[i] = 1
            else:
                vecindario[i] = 0

            vecindarios.append(  vecindario  )


        return vecindarios

    # Se inicializa la mejor_solucion y mejor_valor como una solución vacía con valor 0.
    mejor_solucion = [0] * len(objetos)
    mejor_valor = 0



    # La solucion_actual se inicializa como una solución aleatoria.
    solucion_actual = [random.randint(0, 1) for _ in range(len(objetos))]
    valor_actual = evaluar_solucion(solucion_actual)


    # Definimos como la mejor solución encontrada
    mejor_solucion = solucion_actual.copy()
    mejor_valor = valor_actual

    # Se crea una lista lista_tabu para almacenar las soluciones tabú.
    lista_tabu = []
    sol_t = [] #arreglo con el tiempo que lleva cada solución en LT


    #Se agrega la solución inicial a la lista tabú con Tiempo=1
    lista_tabu.append(  solucion_actual   )
    sol_t.append(1)

    print("Solución inicial: ", solucion_actual)
    print("sol_t:", sol_t)


    #Ciclo principal de búsqueda tabú
    for _ in range(num_iteraciones):

        print("----------------------------------------")

        vecindarios = generar_vecindarios(solucion_actual)
        print("Vecindarios        ", vecindarios)

        # Filtrar vecindarios que no están en la lista tabú
        #Los vecindarios que no están en la lista tabú se almacenan en vecindarios_no_tabu
        vecindarios_no_tabu = [vecindario for vecindario in vecindarios if vecindario not in lista_tabu]
        print("Vecindario no tabu:", vecindarios_no_tabu)

        #se busca el siguiente desplazamiento aleatorio
        num_sol_disponibles = len(vecindarios_no_tabu)
        sig = random.randint(0, num_sol_disponibles-1)
        print("siguiente: ", sig)

        #Si no hay soluciones disponibles, se sale del ciclo
        if not vecindarios_no_tabu:
            print ("No hay soluciones disponibles!!!")
            break



        # La solucion_actual y el valor_actual
        solucion_actual = vecindarios_no_tabu[sig].copy()
        valor_actual = evaluar_solucion(  vecindarios_no_tabu[sig]  )

        # Actualizar la mejor solución encontrada
        if valor_actual > mejor_valor:
            mejor_solucion = solucion_actual.copy()
            mejor_valor = valor_actual



        #eliminamos las soluciones que ya cumplieron el tiempo en la LT
        no_elem_LT = len(sol_t )
        eliminados = 0

        for k in range ( no_elem_LT ):

            if(  k < ( no_elem_LT - eliminados )  ):
                if (sol_t[k] == tiempo_listaT):
                    print("Eliminado: ", lista_tabu[k])
                    sol_t.pop(k)
                    lista_tabu.pop(k)
                    eliminados += 1



        #incrementamos el tiempo de los elementos de la LT
        sol_t = [numero+1 for numero in sol_t]



        #sol_tabu = solucion_actual
        lista_tabu.append(  solucion_actual   )
        sol_t.append(1)
        print("Lista tabu:", lista_tabu)
        print("sol_t:", sol_t)





    print("\n################################")
    print("Mejor solución encontrada:", mejor_solucion)
    print("Valor:", mejor_valor)


    return mejor_solucion



#Objeto 1: valor 60, peso 10
#Objeto 2: valor 100, peso 20
#Objeto 3: valor 120, peso 30
#Objeto 4: valor 40, peso 5
#Objeto 5: valor 50, peso 15

#objetos = [(60, 10), (100, 20), (120, 30), (40, 5), (50, 15)]
objetos = [(6, 6), (10, 10), (12, 12), (5, 5), (7, 7)]
capacidad_maxima = 25
num_iteraciones = 100
t=5 #tiempo de duración en lista tabú (LT)



mejor_solucion = tabu_search(objetos, capacidad_maxima,  num_iteraciones, t)
```
